```python
regualr_sample=regualr_transction.sample(n=499)
```

```python
fraud_data=pd.concat([regualr_sample,fraud_transction])
```

```python
X=fraud_data.drop(columns='Class',axis=1)
```

```python
y= fraud_data['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=1)
```

```python
Model=LogisticRegression()
```

```python
Model.fit(X_train,y_train)
```

```python
X_train_predection=Model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
training_accuracy=accuracy_score(X_train_predection,y_train)*100
```

An f-string allows you to embed expressions, such as variables or calculations, directly inside a string by placing them within curly braces {}.

```python
print(f"Training Data Accuracy: {training_accuracy}%")
```

```python
X_test_predection=Model.predict(X_test)
```

```python
testing_accuracy=accuracy_score(X_test_predection,y_test)*100
```

```python
print(f"Testing_Accuracy:{testing_accuracy}%")
```
