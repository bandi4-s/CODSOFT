```python
sns.pairplot(Data_Set,x_vars=['TV','Radio','Newspaper'],y_vars='Sales',kind='scatter')
plt.show
```

```python
sns.pairplot(Data_Set,x_vars=['TV','Radio','Newspaper'],y_vars='Sales',kind='kde')
plt.show
```

```python
Data_Set.plot.hist(bins=10,orientation='horizontal',color='black')
plt.xlabel('Frequency')
plt.ylabel('TV Advertising Spend')
plt.title ('Histogram of TV Advertising Spend')
```

```python
Data_Set['Radio'].plot.hist(bins=10,color='Skyblue',orientation='horizontal')
plt.xlabel('Radio')
plt.ylabel('TV Advertising Spend')
plt.title('Radio vs TV Advertising Spend ')
plt.show
```

```python
Data_Set['Newspaper'].plot.hist(bins=10,color='black',orientation='horizontal')
plt.xlabel('Newspaper')
plt.ylabel('adv Spend')
plt.title('Newspaper bs adv Spend')
```

```python
sns.heatmap(Data_Set.corr(),annot=True)
plt.show
```
